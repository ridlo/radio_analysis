```python
from astroquery.irsa import Irsa
import astropy.units as u
from astropy import coordinates
```

```python
Irsa.list_catalogs()
```

```python
co = coordinates.SkyCoord(ra=65.81583637500, dec=-1.342518194444, unit=(u.deg, u.deg), frame='fk5', equinox="J2000.0")
```

```python
table = Irsa.query_region(co, catalog="fp_psc", spatial='Box', width=58 * u.arcsec)
```
