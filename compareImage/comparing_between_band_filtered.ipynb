```python
import os
import numpy as np
from astropy import units as u 
from astropy import coordinates
from astropy.io import fits
import matplotlib.pyplot as plt
import aplpy

%matplotlib inline
```

```python
B3 = './images/J0423-0120_B3_filtered.fits'
rmsB3 = 3.5E-05

B6 = './images/J0423-0120_B6_filtered.fits'
rmsB6 = 3.4E-05

B7 = './images/J0423-0120_B7_filtered.fits'
rmsB7 = 5.3E-05

hdu_list = fits.open(B7)
hdu_list.info()
# hdu_list = fits.open(B6)
# hdu_list.info()
# hdu_list = fits.open(B7)
# hdu_list.info()
```

```python
hdu_list[0].header
```

```python
multp = np.array([2.5, 3, 3.5, 4.5, 5, 5.5, 6, 10, 100, 1000])
lvlB3 = rmsB3*multp
lvlB6 = rmsB6*multp
lvlB7 = rmsB7*multp
```

```python
fig = aplpy.FITSFigure(B7, subplot=(3, 1, 1))
fig.show_contour(B7, colors='red', levels=lvlB7, alpha=0.5)
fig.add_beam()
fig.beam.set_color('black')
fig.add_scalebar(1 * u.arcsecond)
fig.scalebar.set_label('1 arcsec')
fig.scalebar.set_color("black")
fig.show_colorscale(cmap="Greys_r")

fig = aplpy.FITSFigure(B6, subplot=(3, 1, 2))
fig.show_contour(B6, colors='red', levels=lvlB6, alpha=0.5)
fig.add_beam()
fig.beam.set_color('black')
fig.add_scalebar(1 * u.arcsecond)
fig.scalebar.set_label('1 arcsec')
fig.scalebar.set_color("black")
fig.show_colorscale(cmap="Greys_r")

fig = aplpy.FITSFigure(B7, subplot=(3, 1, 3))
fig.show_contour(B7, colors='red', levels=lvlB3, alpha=0.5)
fig.add_beam()
fig.beam.set_color('black')
fig.add_scalebar(1 * u.arcsecond)
fig.scalebar.set_label('1 arcsec')
fig.scalebar.set_color("black")
fig.show_colorscale(cmap="Greys_r")
```

### at Band7

```python
fig = aplpy.FITSFigure(B7)
fig.add_beam()
fig.beam.set_color('black')

fig.show_contour(B3, colors='blue', levels=lvlB3, alpha=0.5)#cmap="viridis")
fig.show_contour(B6, colors='green', levels=lvlB6, alpha=0.5)#cmap="autumn")
fig.show_contour(B7, colors='red', levels=lvlB7, alpha=0.5)#cmap="gray")

fig.add_scalebar(1 * u.arcsecond)
fig.scalebar.set_label('1 arcsec')
fig.scalebar.set_color("black")
```
