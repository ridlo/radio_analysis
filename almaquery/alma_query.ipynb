```python
tree = ET.parse("xmldata.xml")
```

```python
root = tree.getroot()
```

```python
print "Number of projects: ", len(root[0][0][0][2][36][0])
```

```python
projects = []

for i, proj in enumerate(root[0][0][0][2][36][0]):
    projects.append([])
    for data in proj:
        if data.text is None:
            projects[i].append('None')
        else:
            projects[i].append(data.text)
```
