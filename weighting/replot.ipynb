```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.loadtxt("onlythebest_B3.txt")
```

```python
pars = data.T
iteration = np.arange(len(pars[0]))

fig, axs = plt.subplots(8,3, figsize=(12,24), sharex=True, sharey=True)
axs = axs.flat

for i, par in enumerate(pars):
    axs[i].plot(iteration, par, 'k.-', alpha=0.3)
    axs[i].set_ylim([-0.1,4.1])
    axs[i].set_title("MS " + str(i+2))
    fig.tight_layout()
```

```python
data = np.loadtxt("log_B3.txt")
```

```python
gen = data[0]
avg = data[2]
min_ = data[3]
max_ = data[4]
```

```python
plt.plot(gen, avg, label="average")
plt.plot(gen, min_, label="minimum")
plt.plot(gen, max_, label="maximum")
plt.xlabel("Generation")
plt.ylabel("Fitness (rms)")
plt.legend(loc="upper right")
plt.show()
```
