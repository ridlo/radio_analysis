```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fringe(uv, x, y):
    res = 0.0
    for i,u in enumerate(uv): 
        res += np.cos(2*np.pi*(uv[i][0]*x/206265 + uv[i][1]*y/206265)) # in arcsec
        
    return res
```

```python
def plot_uv(uv):
    # plot uv-coverage
    plt.plot(uv[:,0], uv[:,1], 'b.')
    plt.plot(-uv[:,0], -uv[:,1], 'b.')
    plt.xlim([-500000, 500000]); plt.ylim([-500000, 500000])
    plt.xlabel("u"); plt.ylabel("v")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
def plot_dirtybeam(uv, n=256, cmap='hot'):
    # plot dirty beam or PSF
    x = np.linspace(-5, 5, n) # in arcsec
    y = np.linspace(-5, 5, n)
    X, Y = np.meshgrid(x, y)

    plt.contourf(X, Y, fringe(uv, X, Y), 8, alpha=.75, cmap=cmap)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

### 2 antenna

```python
uv = np.array([[150000, 300000]])
```

```python
plot_uv(uv)
plot_dirtybeam(uv)
```

### 5 antenna

```python
uv = np.array([[150000, 300000], [50000, -150000], [200000, 200000], [75000, 200000], [200000, -50000]])
```

```python
plot_uv(uv)
plot_dirtybeam(uv)
```

### Random normal
more short spacing

```python
uv = 150000*np.random.randn(300,2) # normal dist
```

```python
plot_uv(uv)
plot_dirtybeam(uv)
```

### Random uniform
more long baseline / large spacing

```python
uv = -500000 + 1000000*np.random.rand(300,2) # uniform dist
```

```python
plot_uv(uv)
plot_dirtybeam(uv)
```
